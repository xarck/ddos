```python
# Check for missing values
missing_values = df1.isnull().sum().sum()
print(f"The dataset contains {missing_values} missing values.")
```

```python
# Remove instances with missing class labels
df1 = df1[~df1[' Label'].isnull()]
```

```python
def correlation(dataset, threshold):
    col_corr = set()  
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
              colname = corr_matrix.columns[i]                  
              col_corr.add(colname)
    return col_corr

corr_features = correlation(df1.iloc[:,:-2], 0.75)
corr_features
```

```python
df1.drop(corr_features,axis=1,inplace=True)
```
